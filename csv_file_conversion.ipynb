```python
from PIL import Image
import numpy as np
import sys
import os
import csv
```

```python
def resize_images(directory, output_size):

    output_directory = os.path.join(directory, 'resized')
    os.makedirs(output_directory, exist_ok=True)
    
    for filename in os.listdir(directory):
        try:
            with Image.open(os.path.join(directory, filename)) as img:
                resized_img = img.resize(output_size)
                
                resized_img.save(os.path.join(output_directory, filename))
                print(f"{filename} resized successfully.")
        except Exception as e:
            print(f"Error resizing {filename}: {e}")

directory = r'C:\Users\tharu\623P\Chatter_free'
output_size = (640, 480)  

# Resize images
resize_images(directory, output_size)
```
